## 3. æ ¸å¿ƒå®ç°

### 3.1 å®Œæ•´çš„RAGç³»ç»Ÿ

```python
import numpy as np
from typing import List, Dict, Any, Optional
from dataclasses import dataclass

@dataclass
class RAGConfig:
    """
    RAGç³»ç»Ÿé…ç½®
    """
    # åˆ†å—é…ç½®
    chunk_size: int = 512
    chunk_overlap: int = 50
    
    # æ£€ç´¢é…ç½®
    top_k: int = 5
    use_hybrid_search: bool = True
    use_reranking: bool = True
    
    # æŸ¥è¯¢å¢å¼ºé…ç½®
    use_hyde: bool = True
    use_query_rewrite: bool = True
    
    # ç¼“å­˜é…ç½®
    enable_cache: bool = True
    cache_ttl: int = 1800  # 30åˆ†é’Ÿ

class InteliKBV2:
    """
    InteliKB v2.0 - å®Œæ•´çš„ä¼˜åŒ–RAGç³»ç»Ÿ
    """
    
    def __init__(self, config: RAGConfig = None):
        self.config = config or RAGConfig()
        self.documents = []
        self.cache = {}  # ç®€åŒ–çš„ç¼“å­˜
        
        print("InteliKB v2.0 åˆå§‹åŒ–å®Œæˆ")
        self._print_config()
    
    def _print_config(self):
        """æ‰“å°é…ç½®"""
        print("\nç³»ç»Ÿé…ç½®ï¼š")
        print("-" * 40)
        print(f"åˆ†å—å¤§å°: {self.config.chunk_size}")
        print(f"æ··åˆæ£€ç´¢: {self.config.use_hybrid_search}")
        print(f"é‡æ’åº: {self.config.use_reranking}")
        print(f"HyDE: {self.config.use_hyde}")
        print(f"æŸ¥è¯¢é‡å†™: {self.config.use_query_rewrite}")
        print(f"ç¼“å­˜: {self.config.enable_cache}")
        print()
    
    def add_documents(self, docs: List[str]):
        """
        æ·»åŠ æ–‡æ¡£
        """
        self.documents.extend(docs)
        print(f"å·²æ·»åŠ  {len(docs)} ä¸ªæ–‡æ¡£")
    
    def query(self, query: str) -> Dict[str, Any]:
        """
        æŸ¥è¯¢æ¥å£
        """
        print(f"\n{'='*60}")
        print(f"æŸ¥è¯¢: {query}")
        print(f"{'='*60}")
        
        # æ­¥éª¤1ï¼šæŸ¥è¯¢å¢å¼º
        enhanced_query = self._enhance_query(query)
        
        # æ­¥éª¤2ï¼šæ£€ç´¢
        retrieved_docs = self._retrieve(enhanced_query)
        
        # æ­¥éª¤3ï¼šé‡æ’åº
        if self.config.use_reranking:
            retrieved_docs = self._rerank(query, retrieved_docs)
        
        # æ­¥éª¤4ï¼šç”Ÿæˆç­”æ¡ˆ
        answer = self._generate(query, retrieved_docs)
        
        return {
            "query": query,
            "answer": answer,
            "sources": [doc[:50] + "..." for doc in retrieved_docs[:3]]
        }
    
    def _enhance_query(self, query: str) -> str:
        """
        æŸ¥è¯¢å¢å¼º
        """
        enhanced = query
        
        if self.config.use_query_rewrite:
            # ç®€åŒ–ï¼šæ·»åŠ ç–‘é—®è¯
            if not query.startswith(('ä»€ä¹ˆ', 'å¦‚ä½•', 'æ€ä¹ˆ', 'ä¸ºä»€ä¹ˆ')):
                enhanced = 'å¦‚ä½•' + enhanced
            print(f"âœ“ æŸ¥è¯¢é‡å†™: {query} -> {enhanced}")
        
        if self.config.use_hyde:
            # HyDEï¼šç”Ÿæˆå‡è®¾ç­”æ¡ˆï¼ˆç®€åŒ–ï¼‰
            hypothetical = f"å…³äº{enhanced}çš„è¯¦ç»†è§£é‡Šï¼ŒåŒ…æ‹¬å®šä¹‰ã€æ–¹æ³•å’Œåº”ç”¨"
            print(f"âœ“ HyDE: ç”Ÿæˆå‡è®¾ç­”æ¡ˆ")
            # å®é™…åº”ç”¨ä¸­ä½¿ç”¨å‡è®¾ç­”æ¡ˆæ£€ç´¢
        
        return enhanced
    
    def _retrieve(self, query: str, top_k: int = None) -> List[str]:
        """
        æ£€ç´¢ï¼ˆç®€åŒ–å®ç°ï¼‰
        """
        top_k = top_k or self.config.top_k
        
        # æ£€æŸ¥ç¼“å­˜
        cache_key = f"{query}:{top_k}"
        if self.config.enable_cache and cache_key in self.cache:
            print(f"âœ“ ç¼“å­˜å‘½ä¸­")
            return self.cache[cache_key]
        
        # ç®€åŒ–ï¼šå…³é”®è¯åŒ¹é…
        query_lower = query.lower()
        results = []
        
        for doc in self.documents:
            score = sum(1 for word in query_lower.split() if word in doc.lower())
            if score > 0:
                results.append((doc, score))
        
        # æ’åº
        results.sort(key=lambda x: x[1], reverse=True)
        retrieved = [doc for doc, _ in results[:top_k]]
        
        # ç¼“å­˜ç»“æœ
        if self.config.enable_cache:
            self.cache[cache_key] = retrieved
        
        print(f"âœ“ æ£€ç´¢åˆ° {len(retrieved)} ä¸ªæ–‡æ¡£")
        return retrieved
    
    def _rerank(self, query: str, docs: List[str]) -> List[str]:
        """
        é‡æ’åºï¼ˆç®€åŒ–å®ç°ï¼‰
        """
        print(f"âœ“ é‡æ’åº {len(docs)} ä¸ªæ–‡æ¡£")
        # ç®€åŒ–ï¼šåŸºäºæ›´ç²¾ç¡®çš„åŒ¹é…
        return docs  # å®é™…åº”ç”¨ä¸­ä½¿ç”¨CrossEncoder
    
    def _generate(self, query: str, docs: List[str]) -> str:
        """
        ç”Ÿæˆç­”æ¡ˆï¼ˆç®€åŒ–ï¼‰
        """
        if not docs:
            return "æŠ±æ­‰ï¼ŒçŸ¥è¯†åº“ä¸­æ²¡æœ‰æ‰¾åˆ°ç›¸å…³ä¿¡æ¯ã€‚"
        
        # ç®€åŒ–ï¼šåŸºäºæ¨¡æ¿ç”Ÿæˆ
        context = " ".join(docs[:2])
        return f"æ ¹æ®æ£€ç´¢åˆ°çš„ä¿¡æ¯ï¼Œ{context[:100]}..."

# åˆ›å»ºç³»ç»Ÿ
print("åˆ›å»ºInteliKB v2.0ç³»ç»Ÿ...\n")
rag_system = InteliKBV2()
```
